## Stored target user's tweets and do Sentiment analyst ( maximum 3200 tweets per request)

```
count=0
for status in tweepy.Cursor(api.user_timeline, screen_name='@realDonaldTrump', tweet_mode="extended").items():
    ## Print tweets
    print(f"{status.user.name} No.{count} at {status.created_at} said {status.full_text}")
    
    ## Start to analyst with textblob
    tweets_analysis = TextBlob(status.full_text)
    print(tweets_analysis.sentiment)
    if tweets_analysis.sentiment[0]>0:
      print ('Positive')
    else:
      print ('Negative')
    print("")
    count=count+1
    
    #print(status.full_text)  ##print in text
    
    
```
